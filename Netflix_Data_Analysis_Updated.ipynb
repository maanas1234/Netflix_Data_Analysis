```python
def categorize(df,col, labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]

    df[col]=pd.cut(df[col],bins=edges, labels=labels, include_lowest =True, duplicates='drop')

    return df
```
